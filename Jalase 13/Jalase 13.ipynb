```python
class IceCreamShop:
    def __init__(self):
        self.capacity = 200_000  # ظرفیت انبار بر حسب گرم
        self.inventory = {
            "شکلاتی": {"weight": 100, "price": 10, "stock": 0},
            "میوه‌ای": {"weight": 55, "price": 7, "stock": 0},
            "موزیکال": {"weight": 200, "price": 15, "stock": 0, "songs": ["آهنگ۱", "آهنگ۲", "آهنگ۳"]},
            "یخی": {"weight": 40, "price": 3, "stock": 0},
        }

    def total_weight(self):
        # محاسبه وزن کل بستنی‌های موجود در انبار
        return sum(item["weight"] * item["stock"] for item in self.inventory.values())

    def add_stock(self, ice_cream_type, quantity):
        # اضافه کردن موجودی به انبار
        if ice_cream_type not in self.inventory:
            print("نوع بستنی اشتباه است!")
            return

        required_space = self.inventory[ice_cream_type]["weight"] * quantity
        if self.total_weight() + required_space > self.capacity:
            print("ظرفیت انبار کافی نیست!")
        else:
            self.inventory[ice_cream_type]["stock"] += quantity
            print(f"{quantity} بستنی {ice_cream_type} به انبار اضافه شد.")

    def sell(self, ice_cream_type, quantity):
        # فروش بستنی
        if ice_cream_type not in self.inventory:
            print("نوع بستنی اشتباه است!")
            return

        if ice_cream_type == "موزیکال" and quantity % 2 != 0:
            print("بستنی موزیکال فقط به صورت جفتی فروخته می‌شود.")
            return

        required_stock = (
            quantity if ice_cream_type != "موزیکال" else quantity * 2
        )  # برای موزیکال دو برابر حساب می‌شود
        if self.inventory[ice_cream_type]["stock"] < required_stock:
            print("موجودی کافی نیست!")
        else:
            self.inventory[ice_cream_type]["stock"] -= required_stock
            income = quantity * self.inventory[ice_cream_type]["price"]
            profit = income * 0.2
            print(
                f"{quantity} بستنی {ice_cream_type} فروخته شد. درآمد: {income} هزار تومان، سود: {profit:.2f} هزار تومان"
            )

            # پخش آهنگ برای بستنی موزیکال
            if ice_cream_type == "موزیکال":
                song_index = required_stock % len(
                    self.inventory["موزیکال"]["songs"]
                )  # انتخاب آهنگ
                print(f"در حال پخش {self.inventory['موزیکال']['songs'][song_index]}")

    def check_inventory(self):
        # نمایش موجودی انبار
        print("موجودی انبار:")
        for name, info in self.inventory.items():
            print(f"{name}: {info['stock']} عدد")


# اجرای برنامه
shop = IceCreamShop()

# افزودن موجودی به انبار
shop.add_stock("شکلاتی", 50)  # اضافه کردن 50 بستنی شکلاتی
shop.add_stock("میوه‌ای", 100)  # اضافه کردن 100 بستنی میوه‌ای
shop.add_stock("موزیکال", 20)  # اضافه کردن 20 بستنی موزیکال
shop.add_stock("یخی", 200)  # اضافه کردن 200 بستنی یخی

# چک کردن موجودی
shop.check_inventory()

# فروش بستنی
shop.sell("شکلاتی", 10)  # فروش 10 بستنی شکلاتی
shop.sell("موزیکال", 4)  # فروش 4 بستنی موزیکال (8 عدد فروخته می‌شود)
shop.sell("یخی", 50)  # فروش 50 بستنی یخی

# چک کردن موجودی دوباره
shop.check_inventory()
```
